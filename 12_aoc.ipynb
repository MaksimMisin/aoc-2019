```python
def tot_e(pos, vel):
    return (np.abs(pos).sum(axis=1) * np.abs(vel).sum(axis=1)).sum()
```

```python
mat = []
for l in data.splitlines():
    l = l.split(',')
    nums  = [int(p.split('=')[1].strip('>')) for p in l]
    mat.append(nums)
pos = np.asarray(mat)
```

```python
pos
```

```python
vel = np.zeros_like(pos, dtype=np.int)
```

```python
vel
```

```python
poss = [pos.copy()]
vels = [vel.copy()]
enes = []
for it in range(150000):
    for i in range(len(pos)):
        m1 = pos[i]
        for j in range(i+1, len(pos)):
            m2 = pos[j]
            #gravity
            m2_v = -(m2 > m1).astype(np.int) + (m2 < m1).astype(np.int)
            m1_v = -m2_v
            vel[j] += m2_v
            vel[i] += m1_v

    for i, m1 in enumerate(pos):
        #position
        pos[i] += vel[i]
    poss.append(pos.copy())
    vels.append(vel.copy())
    #enes.append(tot_e(pos, vel))
```
